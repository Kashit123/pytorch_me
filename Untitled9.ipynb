```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

```python
# check device
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(DEVICE)
```

```python
transforms = transforms.Compose([transforms.Resize((32, 32)),
                                 transforms.ToTensor()])

train_dataset = datasets.MNIST(root='mnist_data', 
                               train=True, 
                               transform=transforms,
                               download=True)

valid_dataset = datasets.MNIST(root='mnist_data', 
                               train=False, 
                               transform=transforms)
```

```python
BATCH_SIZE = 64
train_loader = DataLoader(dataset=train_dataset, 
                          batch_size=BATCH_SIZE, 
                          shuffle=True)

valid_loader = DataLoader(dataset=valid_dataset, 
                          batch_size=BATCH_SIZE, 
                          shuffle=False)
```

```python
class LeNet(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.conv1= nn.Conv2d(1, 6, 5)
        self.conv2= nn.Conv2d(6, 16, 5)
        self.conv3= nn.Conv2d(16, 120, 5)
        self.pool= nn.AvgPool2d(2,2)
        self.l1= nn.Linear(120, 84)
        self.l2= nn.Linear(84,10)
        self.tanh= nn.Tanh()
        self.out= nn.Softmax(dim=1)
    def forward(self, image):
        out= self.tanh(self.conv1(image))
        out= self.pool(out)
        out= self.tanh(self.conv2(out))
        out= self.pool(out)
        out= self.tanh(self.conv3(out))
        out= torch.flatten(out,1)
        out= self.tanh(self.l1(out))
        out= self.out(self.l2(out))
        return out
```

```python
Net= LeNet()
loss_function= nn.CrossEntropyLoss()
optimizer= torch.optim.Adam(Net.parameters(), lr= 0.01)
print(Net)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):

        # Compute prediction error
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch *len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_loader, Net, loss_function, optimizer)
    test(valid_loader, Net, loss_function)
print("Done!")
```
